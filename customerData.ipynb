```python
import pandas as pd
df = pd.read_csv('customer.csv')
print(df)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace = True)
```

```python
print(df.info())
```

```python
df['Registration_Date'] = pd.to_datetime(df['Registration_Date'],format='mixed')
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df['age'] = df['age'].astype(int)
```

```python
df
```

```python
df.columns = df.columns.str.lower()
```

```python
df
```

```python
df['country name'] = df['country name'].replace({'USA':'United States'})
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.loc[:,'gender'] = df['gender'].replace({'f':'female','F':'female'})
```

```python
df
```

```python
df.to_csv("customerDataClean.csv",index=False)
```
